```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import math as m
data_is =  pd.read_csv("NGC2516_isochrone_1.2e8.txt", delimiter=' ')
data_is2 = pd.read_csv("NGC2516_isochrone_2.0e8.txt", delimiter=' ')
data_is3 = pd.read_csv("NGC2516_isochrone_0.7e8.txt", delimiter=' ')

Numx, Numy, x_name,y_name= 21, 15, 'BP-RP, mag','G, mag'

data = pd.read_csv('NGC2516_100m_selected_parallax_0.2_pm_inRA_1.5_pm_inDEC_1.5_CMD_cutted_18_CMD_separated_singles.txt', delimiter=';', header=None)
data.rename(columns = {Numx : x_name, Numy : y_name}, inplace=True)
data = data.apply(pd.to_numeric, errors='coerce')
```

```python
import math as m

fig, ax = plt.subplots(figsize=(10,8 )) #16:14
ax.scatter (data['BP-RP, mag'] , data['G, mag'] , c ='black', s = 3, label ='stars')

ax.plot(data_is['G_BPmag'] - data_is['G_RPmag'] +3.1*0.13*(1.2 - 0.65), data_is['Gmag'] + 5 * m.log10(410) - 5 + 3.1*0.13*0.86, c ='red', lw = 2, label ='1.2e8') 
ax.plot(data_is2['G_BPmag'] - data_is2['G_RPmag'] +3.1*0.13*(1.2 - 0.65), data_is2['Gmag'] + 5 * m.log10(410) - 5 + 3.1*0.13*0.86, c ='blue', lw = 1, label ='2.0e8')
ax.plot(data_is3['G_BPmag'] - data_is3['G_RPmag'] +3.1*0.13*(1.2 - 0.65), data_is3['Gmag'] + 5 * m.log10(410) - 5 + 3.1*0.13*0.86, c ='green', lw = 1,  label ='0.7e8') 
ax.grid(c='#aaaaaa', ls='--') 
locs, labels = plt.yticks()
locs, labels = plt.xticks()
plt.yticks(np.arange(4, 19, step=1.0))
ax.set_ylim(4, 19) 
ax.set_xlim(-0.5, 3)
ax.invert_yaxis()

plt.xlabel(x_name, size = 15)
plt.ylabel(y_name, size = 15)
ax.tick_params(axis='both', which='major', labelsize=15)
ax.legend(fontsize = 15)
plt.savefig("NGC2516_CMD+isochrone.png", dpi=200)
plt.show()
```
